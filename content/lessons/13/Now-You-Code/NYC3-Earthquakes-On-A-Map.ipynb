How to do it:

- Read the earthquake data from http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/1.0_week.geojson  
- Then plot it on a folium map using CircleMarkers (no need to use Pandas). 
- Make the size of the marker correlate to the magnitude of the earthquake (bigger circle for bigger magnitude) also when you click on the marker you should see the title of the earthquake.

Here's a formula for making your markers more noticeable:
`radius = ( 75 * earthquake_magnitude) ** 2` Feel free to tweak the formula.

```python
#todo write your plan here
# import folium
# Map longitude and latitude (Ely, Nevada)
# requests get url of json file
# For loop to plot all earthquake points on the map
# display condition of Earthquake to map
# render properties and magnitude of Json
# map out information

import folium
import random 
import requests
```

```python
#todo write your code here
ely = (39.2474, 114.8886)
map = folium.Map(location = ely, zoom_start=4)
response = requests.get('http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/1.0_week.geojson')
data= response.json()
for quake in data['features'][:]:
    lng = quake['geometry']['coordinates'][0]
    lat = quake['geometry']['coordinates'][1]
    mag = (75 * quake['properties']['mag'])**2
    title= quake['properties']['title']
    marker = folium.CircleMarker(location =(lat,lng),popup=title, radius =mag)
    map.add_child(marker)
map
```
